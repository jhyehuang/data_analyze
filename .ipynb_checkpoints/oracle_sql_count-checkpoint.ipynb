```python
tables=[]
opers=[]
counts=[]
for i in range(len(cds)):
    oper=cds[i][0].lstrip().split(' ')[0].lstrip().split(' ')[0].lower()
    
    if oper=='select':
        print(cds[i][0])
        table=cds[i][0].split('from')[1].lstrip().split(' ')[0]
        print(cds[i][0],i)
    elif oper=='update':
        table=cds[i][0].lstrip().split(' ')[1]
        #print(oper)
    elif oper=='insert':
        table=cds[i][0].split('into')[1].lstrip().split(' ')[0]
        #print(oper)
    count=cds[i][1]
    date=cds[i][2].split('/')[0]
    tables.append(table)
    opers.append(oper)
    counts.append(count)
```

```python
result={}
for i in range(len(opers)):
    if (tables[i]+'_'+opers[i]) in result.keys():
        result[tables[i]+'_'+opers[i]]+=counts[i]
    else:
        result[tables[i]+'_'+opers[i]]=counts[i]
result=sorted(result.items(),key=lambda d:d[1], reverse = True)
#print(result)
```
